```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show    # Tells Python we will use figure and show from Bokeh
from bokeh.io import output_notebook       # Tells Python we will need the output_notebook function


df = pd.read_csv('AAPL.csv') #Import Apple stocks, read the excel file
dates = pd.to_datetime(df['date']).tolist()  # Business days
opening_prices = df['open'].tolist() #Gets opening prices

# Create a figure
p = figure(title='Apple Opening Prices',x_axis_label='Date', x_axis_type = "datetime", y_axis_label='Opening Price')
p.line(x=dates, y=opening_prices, line_width=2, color='blue', legend_label='Open Price')

# Configure the plot
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Open Price (USD)'
p.legend.location = 'top_left'


# Show the plot in the notebook
output_notebook()
show(p)
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show    # Tells Python we will use figure and show from Bokeh
from bokeh.io import output_notebook       # Tells Python we will need the output_notebook function


df = pd.read_csv('AAPL.csv') #Import Apple stocks, read the excel file
dates = pd.to_datetime(df['date']).tolist()  # Business days use df['date'][200:400] for example to check closer in the graph
opening_prices = df['open'].tolist() #Gets opening prices
closing_prices = df['close'].tolist() #Gets closing prices
highs = df['high'].tolist()
lows = df['low'].tolist()
# Create a figure
p = figure(title='Apple Opening Prices',x_axis_label='Date', x_axis_type="datetime", y_axis_label='Opening price')
p.line(x=dates, y=opening_prices, line_width=2, color='blue', legend_label='Open Price')
p.line(x=dates, y=closing_prices, line_width=2, color='red', legend_label='Close Price')
p.line(x=dates, y=highs, line_width=2, color='green', legend_label='High Price')
p.line(x=dates, y=lows, line_width=2, color='orange', legend_label='Low Price')
# Configure the plot
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Open Price (USD)'
p.legend.location = 'top_left'


# Show the plot in the notebook
output_notebook()
show(p)
```
